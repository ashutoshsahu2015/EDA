## Missing Values

```python
### Check the percentage of Nan  values present in each feature

features_with_nan = [features for features in dataset.columns if dataset[features].isnull().sum()>0]

for feature in features_with_nan:
    print(feature,"has", np.round(dataset[feature].isnull().mean()*100,2), '% of missing values')
```

### Since there are many missing values we need to find the relationship between missing values and the dependent feature i.e. SalesPrice


```python
for feature in features_with_nan:
    data=dataset.copy()

    ## Let's make a variable that indicates 1 if the observation was missiing or zero otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)

    ## Let's plot the distribution of the variable
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing value and the dependent variable is clearly visible. So we need to replace these nan values with something meaningful which can be done in the Feature Engineering section

### Numerical Variable

```python
# List of numerical variables

numerical_features= [feature for feature in dataset.columns if dataset[feature].dtype !='O' ]

print ('Number of Numerical Variables:', len (numerical_features))

## Visualizing the Numerical Variables
dataset[numerical_features].head()
```

### Temporal Variables

From the dataset we have found 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. Further analysis of this can be performed in the Feature Engineering.


```python
## List of variables that contain year information

year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
## Let's explore the content of these year variables

for feature in year_feature:
    print(feature , dataset[feature].unique())
```

```python
## Let us analyze the temporal datetime variables
## We will check if there is any relationship between year the house is sold and the saleprice

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('SalePrice')
plt.title('Year Sold vs SalePrice')
plt.show()
```

```python
year_feature
```

```python
## Here we will compare the difference between all the year feature with SalePrice

for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
    
        ## We will compare the difference between year variable and the year house was sold fr 
        data[feature]= data['YrSold'] - data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

There is a relationship between the year_feature and the target variable i.e. SalePrice. For example recently built houses has high prices which can be understood since it is in good condition and vice versa also i.e. the house was built long back has low SalePrice value

```python
## Numerical features are of two types

## 1. Continuous variable 2. Discrete variable

discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]

print('Discrete Variable Count : {}'.format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Let's find the relationship between the discrete feature and the SalePrice

for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature +" vs SalePrice")
    plt.show()
```

There is a relationship between all the discrete feature and the SalePrice variable

### Continuous Variables

```python
continuous_feature = [feature for feature in numerical_features if feature not in year_feature+discrete_feature+['Id']]

print('Continuous feature count : ', len(continuous_feature))
```

```python
## Let us analyze the continuous values relationship by creating histograms to understand the distibutions

for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
## We will be using logarithm transformation

for feature in continuous_feature:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature + " vs SalePrice")
        plt.show()
```

### Outliers

```python
for feature in continuous_feature:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.xlabel(feature)
        plt.show()
```
